```python
import pandas as pd

# Load dataset
df = pd.read_csv('amazon.csv')

# Convert rating column to numeric (errors='coerce' turns invalid values into NaN)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows where rating is NaN (optional, if there are invalid ratings)
df = df.dropna(subset=['rating'])

# Compute average rating per user
user_avg_rating = df.groupby('user_id')['rating'].mean().reset_index()
user_avg_rating.rename(columns={'rating': 'avg_rating'}, inplace=True)

# Show first 10 users
print(user_avg_rating.head(10))

# Optional: stats
print(user_avg_rating['avg_rating'].describe())
```

```python
import matplotlib.pyplot as plt

plt.hist(user_avg_rating['avg_rating'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Average Ratings per User')
plt.xlabel('Average Rating')
plt.ylabel('Number of Users')
plt.show()
```

```python
user_review_count = df.groupby('user_id')['rating'].count().reset_index()
user_review_count.rename(columns={'rating': 'num_reviews'}, inplace=True)
```

```python
most_frequent_product = df['product_id'].mode()[0]
print("Most frequent product:", most_frequent_product)

# If you also want how many times it appears:
count = df['product_id'].value_counts().iloc[0]
print("Number of reviews:", count)
```

```python
# Extract main category (before the first "|")
df['main_category'] = df['category'].str.split('|').str[0]

# Count frequency of each main category
main_cat_counts = df['main_category'].value_counts().reset_index()
main_cat_counts.columns = ['main_category', 'frequency']

print(main_cat_counts)
```

```python
import matplotlib.pyplot as plt

# Extract main category (before the first "|")
df['main_category'] = df['category'].str.split('|').str[0]

# Count frequency of each main category
main_cat_counts = df['main_category'].value_counts()

# Plot top 10 main categories
plt.figure(figsize=(10,6))
main_cat_counts.head(10).plot(kind='bar', color='skyblue', edgecolor='black')

plt.title("Top 10 Main Categories by Frequency")
plt.xlabel("Main Category")
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
